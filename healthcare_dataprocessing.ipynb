```python
import pandas as pd
import numpy as np
```

```python
FILE_NAME = 'C:/Code/Jupyter/DATA/bigdata_healthcare_data.xlsx'

df = pd.read_excel(FILE_NAME, '검진정보_1차')
df2 = pd.read_excel(FILE_NAME, '기본정보').loc[:,['SEX','AGE','SMOKE','DRINK']]


#아래 3항목은 있으면 1, 없으면 결측값이므로 따로 처리
df3 = pd.read_excel(FILE_NAME, '기본정보').loc[:,['HTN','DM','DL']]
df3 = df3.fillna(0)

lifeLog = pd.read_excel(FILE_NAME,'라이프로그').loc[:,['TOTAL_STEP']]
state = pd.read_excel(FILE_NAME, '기본정보').loc[:,'STATE']
gene = pd.read_excel(FILE_NAME,'유전체')
df = pd.concat([df,df2,df3,lifeLog,state,gene],axis=1) #데이터 병합
df = df[df.STATE != 'drop_out'] #삭제 처리될 데이터 처리
df = df[df.DRINK != -9]
df = df[df.SMOKE != -9]
df.drop(['STATE'],axis=1,inplace=True)

#SEX 데이터 문자형 -> 숫자형
df.loc[df['SEX'] == 'M','SEX'] = 1
df.loc[df['SEX'] == 'F','SEX'] = 0
df['SEX'] = df['SEX'].apply(pd.to_numeric)
```

```python


#CRP 데이터 문자형 -> 숫자형
mask = df['CRP_1'].str.contains('<',na=False)
df = df[~mask]
df['CRP_1'] = df['CRP_1'].apply(pd.to_numeric)

#인슐린 데이터 문자형 -> 숫자형
mask = df['Insulin_1'].str.contains('<',na=False)
df = df[~mask]
df['Insulin_1'] = df['Insulin_1'].apply(pd.to_numeric)

df['Insulin_1'] = df['Insulin_1'].apply(pd.to_numeric)
df['Insulin_1'] = df['Insulin_1'].apply(pd.to_numeric)
df['Insulin_1'] = df['Insulin_1'].apply(pd.to_numeric)

df = df.fillna(df.mean()) #결측값 제거
df = df.reset_index(drop=True) #데이터 삭제로 인한 인덱스 초기화
```

```python
mask_Waist = [((df['SEX'] == 1) & (df['Waist_1'] >= 90)) | ((df['SEX'] == 0) & (df['Waist_1'] >= 80))]
mask_TG = [(df['TG_1'] >= 150) | (df['DL'] == 1)]
mask_HDL = [((df['SEX'] == 1) & (df['HDL_1'] < 40)) | ((df['SEX'] == 0) & (df['HDL_1'] < 50)) | (df['DL'] == 1)]
mask_BP = [(df['HTN'] == 1) | ((df['SEX'] == 1) & (df['SBP_1'] >= 130) & (df['DBP_1'] >= 85)) |
           ((df['SEX'] == 0) & (df['SBP_1'] >= 130) & (df['DBP_1'] >= 85))]
mask_GLU0 = [(df['GLU0_1'] >= 100) | (df['DM'] == 1)]

condition = pd.concat([pd.DataFrame(mask_Waist),pd.DataFrame(mask_TG),pd.DataFrame(mask_HDL),
                      pd.DataFrame(mask_BP),pd.DataFrame(mask_GLU0)]).transpose()
condition.columns = ['DWaist', 'DTG', 'DHDL','DBP','DGLU0']
condition['metabolicSyndrome'] = condition.sum(axis=1)
```

```python
df_condition = pd.DataFrame(mask_Waist).transpose()
```

```python
condition
```

```python
df = pd.concat([df,condition],axis=1)
```

```python
#권장 칼로리 계산을 위해 표준체중 계산
df['standard_Weight'] = 0
df['standard_Weight'].mask(df['SEX'] ==1 , ((df['Height_1']/100)**2)*22, inplace = True)
df['standard_Weight'].mask(df['SEX'] ==0 , ((df['Height_1']/100)**2)*21, inplace = True)
```

```python
df.loc[:,['SEX','Height_1','standard_Weight']]
```

```python
#당뇨병 정상,경고,당뇨 판정
df['diabetes'] = 1
df['diabetes'].mask(df['GLU0_1'] < 100 , 0, inplace = True)
df['diabetes'].mask(df['GLU0_1'] > 125 , 2, inplace = True)
```

```python
df.loc[:,['GLU0_1','diabetes']]
```

```python
df['dyslipidemiaHDL'] = 0
df['dyslipidemiaHDL'].mask(df['HDL_1'] >= 60 , 2, inplace = True)

df['dyslipidemiaTG'] = 0
df['dyslipidemiaTG'].mask(df['TG_1'] >= 500 , 3, inplace = True)
df['dyslipidemiaTG'].mask((df['TG_1'] < 500) & (df['TG_1'] >= 200), 2, inplace = True)
df['dyslipidemiaTG'].mask((df['TG_1'] < 200) & (df['TG_1'] >= 150) , 1, inplace = True)

df['dyslipidemia'] = df[['dyslipidemiaTG','dyslipidemiaHDL']].sum(axis=1)
```

```python
df.loc[:,['HDL_1','dyslipidemiaHDL','TG_1','dyslipidemiaTG','dyslipidemia']]
```

```python
df['category'] = 0
df['category'].mask(df['diabetes'] >= 1 , df['category']+1, inplace = True)
df['category'].mask(df['dyslipidemia'] >= 2 , df['category']+2, inplace = True)
df['category'].mask(df['metabolicSyndrome'] >= 3 , df['category']+4, inplace = True)
```

```python
df.loc[:,['diabetes','dyslipidemia','metabolicSyndrome','category']]
```

```python
df.to_csv('C:/Code/Jupyter/DATA/datafile1_fillna.csv', sep=',', na_rep='NaN',encoding='cp949')
```

```python
def getDailyCal(sw,eh):
    if eh == 1:
        return (sw*25,sw*30)
    elif eh == 2:
        return (sw*30,sw*35)
    else:
        return (sw*35,sw*40)
```
